```python
# creating a function for merging all files into one and creating a csv for the file
def merging_files(directory):

    # creating a list comprehend for iterating files in the directory
    files = [file for file in os.listdir(directory)]

    # creating a empty df to store all file's data
    all_months_data = pd.DataFrame()
    
    # concatenating all files into one file
    for file in files:
        df = pd.read_csv(directory + file)
        all_months_data = pd.concat([all_months_data, df])

    # creating csv file
    all_months_data.to_csv("all_data.csv", index=False)
```

```python
merging_files('\\Sales_Data\\')
```

## __Task 3__ : What products are most often sold together?

```python
# creating a df for products sold together
pst = data[data['Order ID'].duplicated(keep=False)]  # keep = false for getting all duplicates
```

```python
# creating a column in df that represents all the products bought based on 'Order ID'
pst['Products Sold Together'] = pst.groupby('Order ID')['Product'].transform(lambda x: ',' .join(x))
```

```python
# now we have got the column for 'Products Sold Together', now let's get rid of duplicates
pst = pst[['Order ID', 'Products Sold Together']].drop_duplicates()
```

```python
pst.head()
```

```python
# using Counter to get combinations of 'Products Sold Together'
from itertools import combinations
from collections import Counter

count = Counter()

for row in pst['Products Sold Together']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 2)))  # 2 to get two combinations

print(count)
```

```python
count.most_common()
```

```python
for key, value in count.most_common(10):
    print(key , ':', value)
```

## __Answer :__  The above products represent the most products bought together followed by denomination
